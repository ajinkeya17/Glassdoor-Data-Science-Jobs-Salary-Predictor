## 1.) Salary Parsing

```python
''' Removing out rows with salary estimates as -1 '''
df = df[df['Salary Estimate']!='-1']
df.info()
```

```python
'''Removing the text from Salary Estimate column'''
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x:x.split('(')[0])
df.head()
```

```python
'''Removing dollar signs and K from the figures'''
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x:x.replace('K','').replace('$',''))
df.head()
```

```python
'''Checking if non-numeric data present in salary column still and yes "Employer Provided Salary" and "Per Hour rate" indicators still present
which we will separate out in distinct columns'''
salaries = df['Salary Estimate'].to_list()
pattern = re.compile(r'([0-9]{1,2}-[0-9]{1,3})(.*)')
for salary in salaries:
    if pattern.search(salary)!=None:
        print(pattern.search(salary).group(2))
     
```

```python
df['Hourly'] = df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
df.head()
```

```python
df['Employer Provided'] = df['Salary Estimate'].apply(lambda x:1 if 'employer provided salary' in x.lower() else 0)
df.head()
```

```python
df[(df['Employer Provided']==1) & (df['Hourly']==1)]
```

```python
'''Retaining only the numeric range values in the salary column'''
pattern = re.compile(r'[0-9]{1,3}-[0-9]{1,3}')
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x:pattern.search(x).group())
df.head(50)
```

```python
'''Create a column for minimum salary i.e. first number of range and a column for max salary i.e. second number of the range'''

df['min salary'] = df['Salary Estimate'].apply(lambda x:x.split('-')[0])
df['max salary'] = df['Salary Estimate'].apply(lambda x:x.split('-')[1])
df.head(50)
```

```python
'''We will use the avg salary as our dependant variable'''
df['avg salary'] = (df['min salary'].astype('int64') + df['max salary'].astype('int64'))/2
df.head()
```

## 2.) Extracting out just the company name

```python
'''If a company has a negative rating then rating is not included in the name and if it is >1 then the rating is also 
included at the end of the name so we remove the last 4 characters in that case'''
df['Company Name Text'] = df.apply(lambda x:x['Company Name'] if x['Rating']<0 else x['Company Name'][:-5], axis=1)
df.head(50)
```

## 3.) Create a State column

```python
df['Job State'] = df['Location'].apply(lambda x:x.split(',')[1] if x.count(',')==1 else x.split(',')[2])  #since some locations are of format district,city,state
df.head()
```

```python
'''to validate if no anomolous data present or not'''
df['Job State'].value_counts()
```

```python
'''New column indicating if the job location is at the headquarters'''
df['same state'] = df.apply(lambda x:1 if x['Location']==x['Headquarters'] else 0, axis=1)
df.head()
```

## 4.) Age of the company in years from present year

```python
df['age of company'] = df['Founded'].apply(lambda x:2022-x if x>1 else x)
df['age of company'].value_counts()
```

## 5.) Parsing out job description descriptors

```python
'''Check presence of keywords like Python, r studio, aws, excel, spark in job description'''
df['python_yn'] = df['Job Description'].apply(lambda x:1 if 'python' in x.lower() else 0)
df[df['python_yn']==1]['Job Description'].iloc[4]
```

```python
df['python_yn'].value_counts()
```

```python
df['R_yn'] = df['Job Description'].apply(lambda x:1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['R_yn'].value_counts()
```

```python
df['aws_yn'] = df['Job Description'].apply(lambda x:1 if 'aws' in x.lower() else 0)
df['aws_yn'].value_counts()
```

```python
df['excel_yn'] = df['Job Description'].apply(lambda x:1 if 'excel' in x.lower() else 0)
df['excel_yn'].value_counts()
```

```python
df['spark_yn'] = df['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
df['spark_yn'].value_counts()
```
